# EXPLORATORY DATA ANALYSIS

## Heatmap to display correlation between given parameters

```python
Premium1 = data.corr()
```

```python
sns.heatmap(Premium1, xticklabels = Premium1.columns, yticklabels = Premium1.columns)
```

## Premium Price VS Any Transplants

```python
plt.figure(figsize = (8, 4))
plt.title('Relation Between Premium Price VS Any Transplants')
plt.xlabel('Number Of Transplants')
plt.ylabel('Average Premium Price')
sns.barplot(x = 'AnyTransplants', y = 'PremiumPrice',ci = None, data = data, palette = 'Set1' )
```

## Average Premium Price Paid By User VS Number Of Major Surgeries Relation

```python
plt.figure(figsize = (12, 6))
plt.title('Relation Between Premium Price Per User vs No of Major Surgeries')
plt.xlabel('Number Of Major Surgeries')
plt.ylabel('Average Premium Price')
sns.barplot(x = 'NumberOfMajorSurgeries', y = 'PremiumPrice', ci = None, data = data, palette = 'rocket')
```

## Premium Price VS Age

```python
plt.figure(figsize = (12, 6))
plt.title('Relation Between Premium Price VS Age')
plt.ylabel('Average Premium Price')
sns.lineplot(x = 'Age', y = 'PremiumPrice',ci = None, data = data)
```

## Average Premium Price vs BloodPressureProblems

```python
plt.figure(figsize = (8, 4))
plt.title('Relation Between Average Premium Price Per User vs BloodPressureProblems')
sns.barplot(x = 'BloodPressureProblems', y = 'PremiumPrice', data = data, palette = 'Set1' )
```

## Premium Price VS Known Allergies

```python
plt.figure(figsize = (8, 4))
plt.title('Relation Between Premium Price VS Known Allergies')
plt.ylabel('Average Premium Price')
sns.barplot(x = 'KnownAllergies', y = 'PremiumPrice', data = data, palette = 'rocket' )
```

## Premium Price Per User vs Diabetes

```python
plt.figure(figsize = (8, 4))
plt.title('Relation Between Premium Price Per User vs Diabetes')
sns.barplot(x = 'Diabetes', y = 'PremiumPrice', data = data, palette = 'Set2' )
```

## Scatterplot History Of Cancer In Family

```python
sns.relplot(x = 'Age', y = 'PremiumPrice', hue = 'HistoryOfCancerInFamily', data = data)
```

## Average Price Per User VS other Parameters comparison

```python
groupDiabetes = data.groupby('Diabetes')['PremiumPrice'].mean()
groupKnownAllergies = data.groupby('KnownAllergies')['PremiumPrice'].mean()
groupBloodPressureProblems = data.groupby('BloodPressureProblems')['PremiumPrice'].mean()
groupAnyTransplants = data.groupby('AnyTransplants')['PremiumPrice'].mean()
groupHistoryOfCancerInFamily = data.groupby('HistoryOfCancerInFamily')['PremiumPrice'].mean()
x = [0, 1]
w = 0.16
bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]
bar5 = [i+w for i in bar4]
plt.figure(figsize = (14, 6))
plt.bar(bar1 ,groupDiabetes, w, label = 'Daibetes')
plt.bar(bar2 ,groupKnownAllergies, w, label = 'Known Allergies' )
plt.bar(bar3 ,groupBloodPressureProblems, w, label = 'Blood Pressure Problems' )
plt.bar(bar4 ,groupAnyTransplants, w, label = 'Any Transplants' )
plt.bar(bar5 ,groupHistoryOfCancerInFamily, w, label = 'HistoryOfCancerInFamily' )
plt.xlabel('Yes / No')
plt.ylabel('Average Premium Price')
plt.title('All Parameters Relation with Premium Price')
plt.xticks(bar1 + (w+0.16), x)
plt.legend()
plt.show()
```
